```python
import numpy as np
import random
import string
```

```python
encode_char = lambda char : bin(ord(char))
decode_char = lambda asscii_binary : chr(int(asscii_binary,2))
encode_string = lambda string : [encode_char(i) for i in string]
decode_string = lambda encoded_string : "".join(decode_char(i) for i in encoded_string)

def generate_chrome(length):
    uppercase_character = string.ascii_uppercase
    lowercase_Character = string.ascii_lowercase
    digits = string.digits
    symbols = string.punctuation
    space = string.whitespace
    tmp = uppercase_character + lowercase_Character + digits + symbols + space

    random_string = "".join(random.choice(tmp) for j in range(length))
    
    return random_string


def generate_population(target_word,size):
    length = len(target_word)

    normal_strings = [generate_chrome(length) for i in range(size)]
    
    population = [encode_string(i) for i in normal_strings]
    return population

generate_population("hello",100)
```

```python
target = "hello"
population = generate_population("hello",1000)

def calculate_fitness(target,chrome):
    score = 0
    encoded_target = encode_string(target)
    for i in range(len(encoded_target)):
        if encoded_target[i] == chrome[i]:
            score+=1

    fitness_value = (score / len(encoded_target)) * 100
    return fitness_value


for chrome in population:
    print(calculate_fitness(target,chrome))

# def fitness_evaluation(population,target):
#     fitness_result = []
#     selection_probability = []
#     for chrome in population:
#         fitness_result.append(calculate_fitness(target,chrome))
#     return fitness_result


# fitness_evaluation(population,target)
```
